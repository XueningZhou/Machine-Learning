### Supervised Learning Algorithms

- k-Nearest Neighbors 

Adjust neighbor numbers in a decision



- Linear Models
y=k1x1 + k2x2 + ... +b. Tend to overfit to your train set, so:

   - L1 Lasso
   
   Adjust alpha to adjust model complexity. Larger alpha ignores more coefficients by set them to be 0.
   - L2 Ridge

   Adjust alpha to restrict the coefficients. Larger alpha pulls coefficients closer to 0.

   - For classification

   Linear SVC and logistic regression: Adjust parameter C. Larger C fits better to train sets, smaller C closer to 0.

   One-vs.-rest for Multiclass classification. Draw a boundary between every two classes and consider the distance between test points and boundaries.
   
- Naive Bayes Classfiers

Set a boundary for every feature, so every class has boundaries for each features. For prediction, compare a data point with the boundaries of every class and find the closest one.


- Decision Trees

Layers of T/F as branches. For continued data, compare to x to a boundary a. 
Adjust depths (layer numbers) for fitting.
Trees cannot predict areas where you don't have training data.



- Ensemble of Decision Trees
   - Random forests

  A collection of trees. Every tree consider different features (Feature number can be adjusted).Compare predicted results of different trees and add different weight according to their accuracy. Final result come from soft votes (average) or hard votes (winner takes all).

   - Gradient boosting trees
   
   Less layer than forests, but add a tree similar to the accurate trees in every layer.

- Kernalized support vector machines (SVM)
Adding nonlinear features to linear models to fit better
- Neural Networks
Next layer depends on multiple nodes in this layer
- Uncertainty estimates from classfiers

```python
from sklearn.model_selection import train_test_split
X, y = mglearn.datasets.make_forge()
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=2)

clf.fit(X_train, y_train)
print(clf.score(X_test, y_test))
```
